```python
from pydoc import describe

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet("../data/raw/yellow_tripdata_2023-01.parquet")
df.head()
```

```python
missing = df.isna().mean().sort_values(ascending=False)
missing[missing>0]
```

```python
missing.head(15).plot(kind="bar")
plt.title("Top 15 Columns by Missing Rate")
plt.show()
```

```python
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"], errors="coerce")
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"], errors="coerce")

df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
df["pickup_day"] = df["tpep_pickup_datetime"].dt.date

df["pickup_hour"].value_counts().sort_index().plot(kind="bar")
plt.title("Trips by Pickup Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Trip Count")
plt.show()
```

```python
df["trip_distance"].describe()
```

```python
sns.histplot(df["trip_distance"], bins=100)
plt.xlim(0, 30)
plt.title("Trip Distance Distribution (030 miles)")
plt.show()
```

```python
sns.boxplot(x=df["fare_amount"])
plt.xlim(0, 100)
plt.title("Fare Amount (0100)")
plt.show()
```

```python
checks = {
    "negative_distanceM": (df["trip_distance"] < 0).mean(),
    "zero_distanceM": (df["trip_distance"] == 0).mean(),
    "negative_fareM": (df["fare_amount"] < 0).mean(),
    "zero_fareM": (df["fare_amount"] == 0).mean(),
    "negative_distance": (df["trip_distance"] < 0).sum(),
    "zero_distance": (df["trip_distance"] == 0).sum(),
    "negative_fare": (df["fare_amount"] < 0).sum(),
    "zero_fare": (df["fare_amount"] == 0).sum(),
}


checks
```

```python
sample = df.sample(20000, random_state=42)  # keep it fast

sns.scatterplot(data=sample, x="trip_distance", y="fare_amount", s=10)
plt.xlim(0, 30)
plt.ylim(0, 150)
plt.title("Fare vs Trip Distance (Sample)")
plt.show()
```

```python
import pandas as pd

checks_df = pd.DataFrame.from_dict(
    {
        "negative_distance": (df["trip_distance"] < 0).sum(),
        "zero_distance": (df["trip_distance"] == 0).sum(),
        "negative_fare": (df["fare_amount"] < 0).sum(),
        "zero_fare": (df["fare_amount"] == 0).sum(),
    },
    orient="index",
    columns=["count"]
)

checks_df
```

```python
total_rows = len(df)

quality_checks = pd.DataFrame({
    "count": [
        (df["trip_distance"] < 0).sum(),
        (df["trip_distance"] == 0).sum(),
        (df["fare_amount"] < 0).sum(),
        (df["fare_amount"] == 0).sum(),
    ],
    "rate": [
        (df["trip_distance"] < 0).mean(),
        (df["trip_distance"] == 0).mean(),
        (df["fare_amount"] < 0).mean(),
        (df["fare_amount"] == 0).mean(),
    ],
}, index=[
    "negative_distance",
    "zero_distance",
    "negative_fare",
    "zero_fare",
])

quality_checks
```
